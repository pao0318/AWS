```python
import pandas as pd
df=pd.read_csv('winequality-red.csv',sep=';')
df.head()
```

```python
df.columns=[label.replace(' ','_') for label in df.columns]
```

```python
df.head()
```

```python
def numeric_to_bucket(df,feature_name):
    median=df[feature_name].median()
    for i,val in enumerate(df[feature_name]):
        if val>=median:
            df.loc[i,'feature_name']='high'
        else:
            df.loc[i,'feature_name']='low'
            
            
        
```

```python
for items in df.columns[:-1]:
    numeric_to_bucket(df,items)
    
```
